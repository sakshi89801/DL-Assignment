Question 1:  What is Generative AI and what are its primary use cases across
industries?


Generative AI is a type of Artificial Intelligence that can create new and original content such as text, images, audio, video, and code.
It learns patterns, structure, and distribution from large datasets.
It is based on models like Transformers (GPT), GANs, and VAEs.

Primary Use Cases:

Text generation: Chatbots like ChatGPT, content writing, summarization
Image generation: AI art tools like DALL·E, MidJourney
Code generation: GitHub Copilot for programming assistance
Healthcare: Medical report generation, drug discovery
Education: Personalized learning systems, AI tutors
Business & marketing: Ad copy, product descriptions
Entertainment: Music composition, game story creation


Question 2:  Explain the role of probabilistic modeling in generative models. How do
these models differ from discriminative models?

Probabilistic modeling focuses on learning data distribution and uncertainty.
It helps models understand how data is generated.

Generative Models:

Learn joint probability P(x, y) or P(x)
Can generate new data similar to training data
Learn how data is formed
Examples: GANs, VAEs, Naive Bayes

Discriminative Models:

Learn conditional probability P(y|x)
Focus only on classification or prediction
Do not generate new data
Examples: Logistic Regression, SVM, Neural Networks

Key Difference:

Generative → models how data is created
Discriminative → separates classes using decision boundary


Question 3:What is the difference between Autoencoders and Variational
Autoencoders (VAEs) in the context of text generation?

Autoencoder:

Neural network used for data compression and reconstruction
Contains encoder (compress) and decoder (rebuild)
Learns fixed latent representation
Deterministic output

Variational Autoencoder (VAE):

Probabilistic version of autoencoder
Encodes input into mean (μ) and variance (σ)
Samples from latent space using probability distribution
Generates new data points

Differences:

Autoencoder: deterministic, only reconstruction
VAE: probabilistic, supports generation
VAE produces smoother latent space
VAE is better for creative tasks like text/image generation


Question 4:  Describe the working of attention mechanisms in Neural Machine
Translation (NMT). Why are they critical?

Attention mechanism allows the model to focus on important words in the input sentence while generating output.
Instead of compressing all information into a single vector, attention uses all encoder outputs.

Working Steps:

Encoder processes input sentence into hidden states
Decoder generates output word-by-word
Attention layer calculates importance scores (weights)
Weighted sum of encoder outputs is passed to decoder

Importance:

Solves long sentence translation problems
Improves accuracy of translation
Helps align words between languages
Provides better context understanding

Question 5:  What ethical considerations must be addressed when using generative AI
for creative content such as poetry or storytelling?

Generative AI raises several ethical challenges due to its ability to create realistic content.

Key Issues:

Bias: Models may reflect gender, racial, or cultural bias
Misinformation: Can generate fake news or deepfakes
Privacy: Training data may contain sensitive personal data
Copyright issues: Generated content may resemble existing works
Job displacement: Automation may reduce human creative jobs
Misuse: Used for scams, phishing, or harmful content

Solutions:

Use clean and diverse datasets
Apply bias detection techniques
Human review systems (human-in-the-loop)
Ethical AI guidelines and regulations
