Question1: Compare and contrast NLTK and spaCy in terms of features, ease of use,
and performance.

NLTK (Natural Language Toolkit)
NLTK is a Python library mainly used for educational and research purposes in NLP.
It provides a large number of linguistic resources such as corpora, word lists, and lexical databases.
It supports tasks like tokenization, stemming, lemmatization, parsing, tagging, and classification.
NLTK is flexible but requires more coding effort for building NLP pipelines.

It is slower in performance, especially for large-scale or real-time applications.
Best suited for learning NLP concepts and academic experimentation.
spaCy
spaCy is a modern, industrial-strength NLP library designed for production use.
It provides pre-trained statistical models for tasks like POS tagging, dependency parsing, and Named Entity Recognition (NER).

It is built for speed and efficiency, making it suitable for large datasets.
spaCy offers a simple, clean, and user-friendly API, reducing coding complexity.
It integrates deep learning-based models for better accuracy in real-world tasks.
Best suited for commercial and real-time NLP applications like chatbots and information extraction systems.

Key Differences
NLTK is academic, spaCy is industrial.
NLTK is slower, spaCy is faster and optimized.
NLTK requires manual pipeline building, spaCy provides prebuilt pipelines.
spaCy is more accurate for modern NLP tasks due to pretrained models.
NLTK is better for learning concepts, spaCy is better for deployment systems.

Question 4: Describe the architecture and functioning of a Recurrent Natural Network
(RNN).

A Recurrent Neural Network (RNN) is a type of neural network designed for sequential data processing.
It is widely used in text processing, speech recognition, and time-series prediction.

Architecture
Consists of input layer, hidden layer, and output layer.
The hidden layer has a loop connection that allows memory of previous inputs.
Each time step takes:
Current input (Xt)
Previous hidden state (Ht-1)

Working
Processes data step-by-step in sequence order.
At each step, it updates its memory using previous hidden state.
Final output depends on entire sequence history.
Mathematical Form
Hidden state update:
Ht = f(Wx·Xt + Wh·Ht-1)
Output:
Yt = g(Why·Ht)

Advantages
Good for sequence modeling tasks like text and speech.
Maintains context from previous inputs.
Suitable for time-dependent data analysis.

Limitations
Suffers from vanishing gradient problem.
Cannot capture long-term dependencies effectively.
Training is slow for long sequences.

Question 5: What is the key difference between LSTM and GRU networks in NLP
applications?

LSTM (Long Short-Term Memory)
LSTM is an advanced type of RNN designed to handle long-term dependencies.
It uses three gates: input gate, forget gate, and output gate.
It has a cell state that stores long-term memory.
More powerful for complex sequence modeling tasks.
Requires more computation and memory.

GRU (Gated Recurrent Unit)
GRU is a simplified version of LSTM.
It uses two gates: reset gate and update gate.
It does not have a separate cell state (uses hidden state only).
It is faster to train and computationally efficient.
Performs almost similar to LSTM in many tasks.

Key Differences
LSTM has 3 gates, GRU has 2 gates.
LSTM uses cell state + hidden state, GRU uses only hidden state.
LSTM is more powerful but slower.
GRU is simpler and faster.
LSTM is better for very long sequences, GRU is better for smaller datasets or faster training needs.

```python
#7
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

# Download required resources
nltk.download('punkt')
nltk.download('punkt_tab')

text = """Natural Language Processing (NLP) is a fascinating field that combines linguistics,
computer science, and artificial intelligence. It enables machines to understand, interpret,
and generate human language. Applications of NLP include chatbots, sentiment analysis, and
machine translation. As technology advances, the role of NLP in modern solutions is becoming
increasingly critical."""

# Tokenization
tokens = word_tokenize(text)

# Frequency Distribution
fdist = FreqDist(tokens)

print("TOKENS:\n", tokens)
print("\nTOP 10 FREQUENT WORDS:\n", fdist.most_common(10))
```

```python
#8
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

# Dataset
texts = [
    "I love this project",
    "This is an amazing experience",
    "I hate waiting in line",
    "This is the worst service",
    "Absolutely fantastic!"
]

labels = [1, 1, 0, 0, 1]

# Tokenization
tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)

# Padding
max_len = max(len(x) for x in sequences)
X = pad_sequences(sequences, maxlen=max_len)

y = np.array(labels)

# Model
model = Sequential()
model.add(Embedding(input_dim=50, output_dim=16, input_length=max_len))
model.add(LSTM(32))
model.add(Dense(1, activation='sigmoid'))

# Compile
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train
model.fit(X, y, epochs=10, verbose=1)

# Prediction
print(model.predict(X))
```
