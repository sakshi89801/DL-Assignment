1: What is the role of filters and feature maps in Convolutional Neural
Network (CNN)?

Filters (Kernels) detect patterns
Filters are small matrices (e.g., 3×3, 5×5) used to scan input images.

Each filter learns specific features like edges, corners, textures, or shapes.
Feature extraction process

When a filter slides over the image, it performs convolution operation.
It multiplies values and sums them to detect presence of a feature.

Feature maps represent detected patterns
Output of convolution is called a feature map.
It shows where a particular feature appears in the image.

Multiple filters → multiple feature maps
Each filter produces one feature map.
CNN learns many filters to detect different patterns simultaneously.

Importance in deep learning
Early layers detect simple features (edges).
Deeper layers detect complex features (faces, objects).

Question 2: Explain the concepts of padding and stride in CNNs(Convolutional Neural
Network). How do they affect the output dimensions of feature maps?

Padding means adding extra pixels (usually zeros) around the image border.

It helps preserve spatial size after convolution.
Types of padding

Valid padding: No padding (output size reduces).
Same padding: Output size remains same as input.
Stride

Stride is the number of pixels the filter moves at each step.
Stride = 1 means move one pixel at a time.
Stride = 2 means skip pixels and move faster.
Effect on output size
Larger stride → smaller feature map.
More padding → larger or same-sized output.

Formula impact
Output size depends on:

	​

+1

where N = input size, F = filter size, P = padding, S = stride.

Question 3: Define receptive field in the context of CNNs. Why is it important for deep
architectures?

Receptive field is the region of input image that affects a neuron’s output.

Local perception
Each neuron sees only a small part of the image, not the whole image.

Increases with depth
Early layers → small receptive field.
Deeper layers → larger receptive field.

Importance in CNN
Helps model understand spatial hierarchy of features.

Enables detection of complex patterns like faces or objects.
Key role

Larger receptive field improves global understanding of images.

Question 4: Discuss how filter size and stride influence the number of parameters in a
CNN.

Filter size affects learnable parameters
Larger filters (e.g., 5×5) have more weights than smaller filters (e.g., 3×3).

More weights = more parameters.
Formula for parameters
Parameters = (filter height × filter width × input channels × number of filters)

Stride does NOT affect parameters directly
Stride changes output size, not weights.

Indirect effect of stride
Higher stride reduces feature map size.
This reduces computation in deeper layers.

Optimization trade-off
Small filters (3×3) are preferred (used in VGG).
They reduce parameters while keeping performance high.

Question 5: Compare and contrast different CNN-based architectures like LeNet,
AlexNet, and VGG in terms of depth, filter sizes, and performance.

LeNet
One of the earliest CNN models.
Shallow network (5–7 layers).
Uses small datasets like MNIST.
Small filters (5×5).

AlexNet
Deeper than LeNet (~8 layers).
Uses ReLU activation and dropout.
Works on large datasets (ImageNet).
Introduced GPU training.

VGG
Very deep network (16–19 layers).
Uses only 3×3 filters.
Simple but powerful architecture.
High computational cost.

Comparison
LeNet → simple, low accuracy, small data.
AlexNet → breakthrough model, improved accuracy.
VGG → very deep, high accuracy but heavy computation.

Performance trend
Depth increases → accuracy improves but computation also increases.

```python
#6
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import mnist

# Load dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize
x_train = x_train.reshape(-1, 28, 28, 1) / 255.0
x_test = x_test.reshape(-1, 28, 28, 1) / 255.0

# Model
model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# Compile
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train
model.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test))

# Evaluate
test_loss, test_acc = model.evaluate(x_test, y_test)
print("Test Accuracy:", test_acc)
```

```python
#7
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import cifar10

# Load dataset
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# Normalize
x_train, x_test = x_train / 255.0, x_test / 255.0

# Model
model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(32,32,3)),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(128, (3,3), activation='relu'),

    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(10, activation='softmax')
])

# Compile
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train
model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test))

# Evaluate
model.evaluate(x_test, y_test)
```

```python
#8
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms

# -----------------------------
# Device (GPU if available)
# -----------------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# -----------------------------
# Transform
# -----------------------------
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

# -----------------------------
# MNIST Dataset
# -----------------------------
train_dataset = torchvision.datasets.MNIST(
    root="./data",
    train=True,
    transform=transform,
    download=True
)

test_dataset = torchvision.datasets.MNIST(
    root="./data",
    train=False,
    transform=transform,
    download=True
)

train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=64, shuffle=False)

# -----------------------------
# CNN Model
# -----------------------------
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()

        self.conv1 = nn.Conv2d(1, 32, 3)
        self.conv2 = nn.Conv2d(32, 64, 3)
        self.pool = nn.MaxPool2d(2, 2)

        self.fc1 = nn.Linear(64 * 5 * 5, 128)
        self.fc2 = nn.Linear(128, 10)

        self.relu = nn.ReLU()

    def forward(self, x):
        x = self.pool(self.relu(self.conv1(x)))
        x = self.pool(self.relu(self.conv2(x)))
        x = x.view(x.size(0), -1)
        x = self.relu(self.fc1(x))
        x = self.fc2(x)
        return x

# -----------------------------
# Model, Loss, Optimizer
# -----------------------------
model = CNN().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# -----------------------------
# Training (FAST VERSION)
# -----------------------------
epochs = 1   # keep small for exam/Colab

for epoch in range(epochs):
    model.train()
    running_loss = 0

    for i, (images, labels) in enumerate(train_loader):
        images, labels = images.to(device), labels.to(device)

        outputs = model(images)
        loss = criterion(outputs, labels)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

        # limit batches for speed
        if i > 100:
            break

    print(f"Epoch [{epoch+1}], Loss: {running_loss:.4f}")

# -----------------------------
# Evaluation
# -----------------------------
model.eval()
correct = 0
total = 0

with torch.no_grad():
    for i, (images, labels) in enumerate(test_loader):
        images, labels = images.to(device), labels.to(device)

        outputs = model(images)
        _, predicted = torch.max(outputs, 1)

        total += labels.size(0)
        correct += (predicted == labels).sum().item()

        if i > 50:   # fast evaluation
            break

accuracy = 100 * correct / total
print("Test Accuracy:", accuracy, "%")

```

```python
#9
import os
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models

# =========================
# 1. LOCAL DATASET PATH
# =========================
dataset_path = "./dataset"   # upload dataset folder in Colab or local system

# =========================
# 2. CHECK DATASET
# =========================
if not os.path.exists(dataset_path):
    raise Exception("Dataset folder not found. Upload dataset in working directory!")

print(" Dataset Found!")

# =========================
# 3. IMAGE PREPROCESSING
# =========================
datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2,
    rotation_range=20,
    zoom_range=0.2,
    horizontal_flip=True
)

# =========================
# 4. TRAIN DATA
# =========================
train_data = datagen.flow_from_directory(
    dataset_path,
    target_size=(150, 150),
    batch_size=32,
    class_mode='binary',
    subset='training'
)

# =========================
# 5. VALIDATION DATA
# =========================
val_data = datagen.flow_from_directory(
    dataset_path,
    target_size=(150, 150),
    batch_size=32,
    class_mode='binary',
    subset='validation'
)

# =========================
# 6. CNN MODEL
# =========================
model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(150,150,3)),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(128, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(1, activation='sigmoid')
])

# =========================
# 7. COMPILE MODEL
# =========================
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# =========================
# 8. TRAIN MODEL
# =========================
model.fit(
    train_data,
    validation_data=val_data,
    epochs=10
)

# =========================
# 9. SAVE MODEL
# =========================
model.save("cnn_model_q9.h5")

print(" Model trained and saved successfully!")
```
