Question 1: What are Large Language Models (LLMs) and how do they function?

Answer:

Large Language Models (LLMs) are advanced AI systems designed to understand and generate human-like text.
They are trained on massive datasets containing books, articles, websites, and code.
LLMs use a neural network architecture called Transformer, which helps them understand context in text.

They work by predicting the next word (token) in a sentence based on previous words.
During training, they learn grammar, facts, reasoning patterns, and language structure.
Examples include GPT, LLaMA, Claude, etc.
They can perform tasks like writing, summarizing, translation, coding, and Q&A.


Question 2: Discuss the impact of LLMs on traditional software development approaches.

Answer:

LLMs reduce the need for manual coding by generating code automatically.
Developers can now use prompt-based development instead of writing full programs.
They speed up software development life cycle (SDLC) by assisting in design, coding, and testing.
Debugging becomes easier as LLMs can suggest fixes for errors.
Traditional rule-based systems are being replaced by AI-driven systems.

However, developers still need to validate AI-generated outputs for correctness and security.
LLMs shift the developer role from “coding everything” to “supervising AI-generated code.”

Question 3: Advantages and limitations of LLMs in real-world applications

Answer:

Advantages:
High-quality text generation and understanding.
Automates repetitive tasks like summarization and report writing.
Supports multiple languages.
Improves productivity in coding and content creation.
Can be integrated into chatbots and virtual assistants.
Limitations:
May generate incorrect or hallucinated information.
Requires high computational resources.
Lacks true understanding or consciousness.
Sensitive to biased training data.
Limited real-time knowledge unless connected to external tools.

Question 5: Compare LangChain and LlamaIndex


LangChain:

Framework for building applications using LLMs.
Focuses on chaining multiple steps (prompt → model → tools → output).
Good for building agents, workflows, and API integrations.
Supports memory, tools, and external APIs.

LlamaIndex:
Framework for connecting LLMs with external data sources.
Focuses on indexing and retrieving documents efficiently.
Used for document-based Q&A systems.
Optimized for retrieval-augmented generation (RAG).

Key Difference:
LangChain = workflow + automation + agents
LlamaIndex = data indexing + retrieval system
